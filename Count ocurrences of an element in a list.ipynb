```python
def count_ocurrences(l, element):
    count = l.count(element)
    return count

#Example
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
element_to_count = 3
ocurrences = count_ocurrences(my_list, element_to_count)
print(f"The element {element_to_count} appears {ocurrences} times in the list.")
```
